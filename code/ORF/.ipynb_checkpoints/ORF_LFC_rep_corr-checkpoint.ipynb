```python
import pandas as pd
from poola import core as pool
import seaborn as sns
import gpplot
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import gpplot as gpp
import re
import statistics 
import scipy as sp
```

```python
def olfactory_gene(OG_read_counts, gene_col_name = 'Construct IDs'):
    #OG_read_counts = orf_rc
    olfactory_gene = [gene for gene in OG_read_counts[gene_col_name].tolist() if gene.startswith('OR')]
    # A nomenclature system has been devised for the olfactory receptor family[33] and is the basis for the 
    #official Human Genome Project (HUGO) symbols for the genes that encode these receptors. 
    #The names of individual olfactory receptor family members are in the format "ORnXm" where:

    # OR is the root name (Olfactory Receptor superfamily)
    # n = an integer representing a family (e.g., 1-56) whose members have greater than 40% sequence identity,
    # X = a single letter (A, B, C, ...) denoting a subfamily (>60% sequence identity), and
    # m = an integer representing an individual family member (isoform).
    # checked X could be mutiple letters

    ofg_l = []
    no_ofg_l = []
    for ofg in olfactory_gene:
        match = re.search('^OR\d{1,2}[A-Z]{1,2}[0-9]?[0-9]', ofg)
        if match:
            ofg_l.append(ofg)
        else:
            no_ofg_l.append(ofg)

    return ofg_l
```

```python
path = ('../manuscript data/Supplementary Data Files/')
orf_rc = pd.read_excel(path + '/Supplementary Data 4.xlsx', 'Read counts', 
                             skiprows=2, names = ['Gene Symbol', 'Meljuso_ORF_DO_RepA_D4', 
                                                  'Meljuso_ORF_DO_RepB_D4','Meljuso_ORF_DO_RepA_D21',
                                                  'Meljuso_ORF_DO_RepB_D21','Meljuso_ORF_Sel_RepA_D21',
                                                  'Meljuso_ORF_Sel_RepB_D21'])

chip = pd.read_excel(path + '/Supplementary Data 4.xlsx', 'Library annotation')

```

```python
orf_rc_meljuso = pd.concat([chip,orf_rc], axis =1)
```

```python
sum(orf_rc_meljuso['Gene'] == orf_rc_meljuso['Gene Symbol']) == len(orf_rc_meljuso)
```

```python
orf_rc_meljuso = orf_rc_meljuso.drop(columns=['Gene'])
```

```python
orf_rc_meljuso
```

```python
olfactory_gene_orf = olfactory_gene(orf_rc_meljuso, gene_col_name = 'Gene Symbol')
```

```python
orf_lognorm = pool.lognorm_columns(reads_df=orf_rc_meljuso, columns=['Meljuso_ORF_DO_RepA_D4', 'Meljuso_ORF_DO_RepB_D4',
                                                             'Meljuso_ORF_DO_RepA_D21','Meljuso_ORF_DO_RepB_D21',
                                                             'Meljuso_ORF_Sel_RepA_D21','Meljuso_ORF_Sel_RepB_D21'])
```

# Calculate LFC

```python
orf_lognorm = orf_lognorm[['Transcript', 'Gene Symbol', 
                           'Meljuso_ORF_DO_RepA_D4','Meljuso_ORF_DO_RepB_D4',
                           'Meljuso_ORF_DO_RepA_D21','Meljuso_ORF_DO_RepB_D21',
                           'Meljuso_ORF_Sel_RepA_D21','Meljuso_ORF_Sel_RepB_D21']]
```

```python
#ORF Dropout
OG_lfc_df_do_a = pool.calculate_lfcs(lognorm_df=orf_lognorm, ref_col='Meljuso_ORF_DO_RepA_D4', 
                             target_cols='Meljuso_ORF_DO_RepA_D21')
OG_lfc_df_do_b = pool.calculate_lfcs(lognorm_df=OG_lfc_df_do_a, ref_col='Meljuso_ORF_DO_RepB_D4', 
                             target_cols='Meljuso_ORF_DO_RepB_D21')
```

```python
#ORF Selumetinib
OG_lfc_df_from_sel_a = pool.calculate_lfcs(lognorm_df=orf_lognorm, ref_col='Meljuso_ORF_DO_RepA_D21', 
                             target_cols='Meljuso_ORF_Sel_RepA_D21')
OG_lfc_df_from_sel_b = pool.calculate_lfcs(lognorm_df=OG_lfc_df_from_sel_a, ref_col='Meljuso_ORF_DO_RepB_D21', 
                             target_cols='Meljuso_ORF_Sel_RepB_D21')
```

# LFC replicates corr


```python
OG_lfc_df_from_sel_b['olfactory_gene'] = OG_lfc_df_from_sel_b['Gene Symbol'].isin(olfactory_gene_orf)
OG_lfc_df_do_b['olfactory_gene'] = OG_lfc_df_do_b['Gene Symbol'].isin(olfactory_gene_orf)
```

```python
plt.subplots(figsize=(5,5))
sns.scatterplot(data=OG_lfc_df_do_b, x='Meljuso_ORF_DO_RepA_D21', y='Meljuso_ORF_DO_RepB_D21', alpha = 0.8,
               color = sns.color_palette('Set2')[2]).set(title='VP64 vs p65 (selumetinib)')
sns.scatterplot(data=OG_lfc_df_do_b[OG_lfc_df_do_b['olfactory_gene'] == True], 
                x='Meljuso_ORF_DO_RepA_D21', y='Meljuso_ORF_DO_RepB_D21', 
                alpha = 0.8,
               color = sns.color_palette('Set2')[1])
sns.despine()
plt.xlabel('ORF Dropout RepA Day21 LFC',fontsize=14)
plt.ylabel('ORF Dropout RepB Day21 LFC',fontsize=14)
plt.title('Viability Replicate Correlation',fontsize=16)
r, p = sp.stats.pearsonr(x=OG_lfc_df_do_b['Meljuso_ORF_DO_RepA_D21'], y=OG_lfc_df_do_b['Meljuso_ORF_DO_RepB_D21'])
ax = plt.gca()
plt.text(.07, 0.92, 'r = {:.2f}'.format(r), transform=ax.transAxes)
```

```python
plt.subplots(figsize=(5,5))
sns.scatterplot(data=OG_lfc_df_from_sel_b,x='Meljuso_ORF_Sel_RepA_D21', y='Meljuso_ORF_Sel_RepB_D21', alpha = 0.8,
               color = sns.color_palette('Set2')[2]).set(title='VP64 vs p65 (selumetinib)')
sns.scatterplot(data=OG_lfc_df_from_sel_b[OG_lfc_df_from_sel_b['olfactory_gene'] == True], 
                x='Meljuso_ORF_Sel_RepA_D21', y='Meljuso_ORF_Sel_RepB_D21', 
                alpha = 0.8,
               color = sns.color_palette('Set2')[1])
sns.despine()
plt.xlabel('ORF Selumetinib RepA Day21 LFC',fontsize=14)
plt.ylabel('ORF Selumetinib RepB Day21 LFC',fontsize=14)
plt.title('Selumetinib Modifier Replicate Correlation',fontsize=16)
r, p = sp.stats.pearsonr(x=OG_lfc_df_from_sel_b['Meljuso_ORF_Sel_RepA_D21'], 
                         y=OG_lfc_df_from_sel_b['Meljuso_ORF_Sel_RepB_D21'])
ax = plt.gca()
plt.text(.07, 0.92, 'r = {:.2f}'.format(r), transform=ax.transAxes)
```
